## 5. Thresholding 

```python
img = cv.imread('figuras/sitio_arqueologico.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,100 ,255,cv.THRESH_BINARY)
plt.imshow( img_thr,cmap='gray')
plt.show()
```

## 6. Resize o cambio de tamaño

```python
img = cv.imread('figuras/sitio_arqueologico.jpg')
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,100 ,255,cv.THRESH_BINARY)
alto=img.shape[0]
ancho=img.shape[1]
ratio=0.02
img_r = cv.resize(img_thr,(int(ancho*ratio),int(alto*ratio)), interpolation=cv.INTER_NEAREST)
plt.imshow( img_r,cmap='gray')
plt.show()
```

## 7. Todas las operaciones

```python
img = cv.imread('figuras/sitio_arqueologico.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(31,31),0)
thr, img_thr= cv.threshold(img_gauss ,100 ,255,cv.THRESH_BINARY)
alto=img.shape[0]
ancho=img.shape[1]
ratio=0.02
img_r = cv.resize(img_thr,(int(ancho*ratio),int(alto*ratio)), interpolation=cv.INTER_NEAREST)

plt.figure(figsize=(10,6))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Imagen original")
plt.subplot(2,3,2)
plt.imshow(img_gray, cmap='gray')
plt.title("Imagen en escala de grises")
plt.subplot(2,3,3)
plt.imshow(img_gauss, cmap='gray')
plt.title("Imagen con filtro Gaussiano")
plt.subplot(2,3,4)
plt.imshow(img_thr, cmap='gray')
plt.title("Imagen con Thresholding")
plt.subplot(2,3,5)
plt.imshow(img_r, cmap='gray')
plt.title("Imagen con el ratio = " + str(ratio))
plt.show()
```

## 1. Mi carretara

```python
#Diferentes operaciones para la imagen carretera.jpg
#abrir una imagen en jupyter con cv2
img = cv.imread('figuras/carretera.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img = cv.imread('figuras/carretera.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)

img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img_gauss = cv.GaussianBlur(img_gray,(3,3),0)
thr, img_thr= cv.threshold(img_gauss ,160 ,255,cv.THRESH_BINARY)
alto=img.shape[0]
ancho=img.shape[1]
ratio=0.018
img_r = cv.resize(img_thr,(int(ancho*ratio),int(alto*ratio)), interpolation=cv.INTER_NEAREST)

plt.figure(figsize=(15,9))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Imagen original")
plt.subplot(2,3,2)
plt.imshow(img_gray,cmap='gray')
plt.title("Imagen en escala de grises")
plt.subplot(2,3,3)
plt.imshow(img_gauss, cmap='gray')
plt.title("Imagen con filtro Gaussiano")
plt.subplot(2,3,4)
plt.imshow(img_thr, cmap='gray')
plt.title("Imagen con Thresholding")
plt.subplot(2,3,5)
plt.imshow(img_r, cmap='gray')
plt.title("Imagen con el ratio = " + str(ratio))
plt.show()
```
