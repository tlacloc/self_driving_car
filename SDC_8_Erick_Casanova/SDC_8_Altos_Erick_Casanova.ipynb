# Cargar librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
stop_cascade = cv.CascadeClassifier('datos/stopsign_classifier.xml')
```

```python
img0 = cv.imread('figuras/imagen_0.jpg')
img1 = cv.imread('figuras/imagen_1.jpg')
img2 = cv.imread('figuras/imagen_2.jpg')
img3 = cv.imread('figuras/imagen_3.jpg')
img4 = cv.imread('figuras/imagen_4.jpg')

#Convertimos a escala de grises la imagen original
img0_grey = cv.cvtColor(img0, cv.COLOR_BGR2GRAY)
img1_grey = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2_grey = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3_grey = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)
img4_grey = cv.cvtColor(img4, cv.COLOR_BGR2GRAY)
```

```python
alto0 = stop_cascade.detectMultiScale(img0_grey, 1.12, 5)
alto1 = stop_cascade.detectMultiScale(img1_grey, 1.12, 5)
alto2 = stop_cascade.detectMultiScale(img2_grey, 1.12, 5)
alto3 = stop_cascade.detectMultiScale(img3_grey, 1.12, 5)
alto4 = stop_cascade.detectMultiScale(img4_grey, 1.12, 5)
```

```python
def detector_altos(alto, img):
    for (x,y,w,h) in alto:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.figure(figsize=(13,8))
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
img0 = detector_altos(alto0, img0)
img1 = detector_altos(alto1, img1)
img2 = detector_altos(alto2, img2)
img3 = detector_altos(alto3, img3)
img4 = detector_altos(alto4, img4)
```

```python
plt.figure(figsize=(17,17))

plt.subplot(2,3,1)
plt.imshow(img0)
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1)
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2)
plt.title("Imagen 2")

plt.subplot(2,3,4)
plt.imshow(img3)
plt.title("Imagen 3")

plt.subplot(2,3,5)
plt.imshow(img4)
plt.title("Imagen 4")

plt.show()
```
