# Cargar librerias

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2 as cv
```

```python
people_cascade = cv.CascadeClassifier('datos/haarcascade_fullbody.xml')
```

```python
img5 = cv.imread('figuras/imagen_5.jpg')
img6 = cv.imread('figuras/imagen_6.jpg')
img7 = cv.imread('figuras/imagen_7.jpg')
img8 = cv.imread('figuras/imagen_8.jpg')
img9 = cv.imread('figuras/imagen_9.jpg')

#Convertimos a escala de grises la imagen original
img5_grey = cv.cvtColor(img5, cv.COLOR_BGR2GRAY)
img6_grey = cv.cvtColor(img6, cv.COLOR_BGR2GRAY)
img7_grey = cv.cvtColor(img7, cv.COLOR_BGR2GRAY)
img8_grey = cv.cvtColor(img8, cv.COLOR_BGR2GRAY)
img9_grey = cv.cvtColor(img9, cv.COLOR_BGR2GRAY)
```

```python
persona5 = people_cascade.detectMultiScale(img5_grey, 1.12, 5)
persona6 = people_cascade.detectMultiScale(img6_grey, 1.12, 5)
persona7 = people_cascade.detectMultiScale(img7_grey, 1.12, 5)
persona8 = people_cascade.detectMultiScale(img8_grey, 1.12, 5)
persona9 = people_cascade.detectMultiScale(img9_grey, 1.12, 5)
```

```python
def detector_personas(persona, img):
    for (x,y,w,h) in persona:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.figure(figsize=(13,8))
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
img5 = detector_personas(persona5, img5)
img6 = detector_personas(persona6, img6)
img7 = detector_personas(persona7, img7)
img8 = detector_personas(persona8, img8)
img9 = detector_personas(persona9, img9)
```

```python
plt.figure(figsize=(17,10))

plt.subplot(2,3,1)
plt.imshow(img5)
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img6)
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img7)
plt.title("Imagen 2")

plt.subplot(2,3,4)
plt.imshow(img8)
plt.title("Imagen 3")

plt.subplot(2,3,5)
plt.imshow(img9)
plt.title("Imagen 4")

plt.show()
```

# Parte B

```python
def people_finder(img):
    #convertir a escala de grises la imagen
    img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
    #encontrar personas en las imágnees
    persona = people_cascade.detectMultiScale(img_grey, 1.12, 5)

    return persona
```

```python
video = cv.VideoCapture("videos/peatones.avi")

cv.startWindowThread()
while(video.isOpened()):
    ret, frame = video.read()
    alto = frame.shape[0]
    ancho = frame.shape[1]
    ratio = 0.8
    frame =cv.resize(frame, ( int(ancho*ratio) , int(alto*ratio) ), interpolation=cv.INTER_NEAREST )
    #print(ret)
    if ret:
        #detecta personas
        people_detected = people_finder(frame)
        #show frames
        for (x,y,w,h) in people_detected:
            cv.rectangle(frame,(x, y),(x+w, y+h), (0,0,255),2)
            
        cv.imshow("video con personas", frame)   
        if cv.waitKey(1) & 0xFF == ord('q'):
            print('q')
            break
    else:
        break

cv.waitKey(5000)
cv.destroyWindow('video_original')
        
cv.waitKey(1)

cv.destroyAllWindows()

video.release()

cv.waitKey(1)
```
