## Caras

```python
#Cargando la imagen
img = cv.imread('figuras/people_0.jpg')
img1 = cv.imread('figuras/people_1.jpg')
img2 = cv.imread('figuras/people_2.jpg')
img3 = cv.imread('figuras/people_3.jpg')
img4 = cv.imread('figuras/people_4.jpg')
```

```python
#Cargando el clasificador
face_cascade = cv.CascadeClassifier('datos/haarcascade_frontalface_default.xml')
```

```python
#Convertimos a escala de grises la imagen original
img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img1_grey = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2_grey = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3_grey = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)
img4_grey = cv.cvtColor(img4, cv.COLOR_BGR2GRAY)

plt.figure(figsize=(13,8))
plt.subplot(2,3,1)
plt.imshow(img_grey, cmap='gray')
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2_grey, cmap='gray')
plt.title("Imagen 3")

plt.subplot(2,3,4)
plt.imshow(img3_grey, cmap='gray')
plt.title("Imagen 4")

plt.subplot(2,3,5)
plt.imshow(img4_grey, cmap='gray')
plt.title("Imagen 5")

plt.show()
```

```python
#Realizamos la búsqueda de rostros en la imagen
# 1.1 y 5 son valores típicos para una buena búsqueda
faces = face_cascade.detectMultiScale(img_grey, 1.12, 5)
faces1 = face_cascade.detectMultiScale(img1_grey, 1.12, 5)
faces2 = face_cascade.detectMultiScale(img2_grey, 1.12, 5)
faces3 = face_cascade.detectMultiScale(img3_grey, 1.12, 5)
faces4 = face_cascade.detectMultiScale(img4_grey, 1.12, 5)

print(faces)
```

```python
def rectangulos_cara(faces, img):
    for (x,y,w,h) in faces:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    plt.figure(figsize=(13,8))
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
#Dibujamos rectángulos sobre la imagen original y mostramos nuevamente la imagen
img = rectangulos_cara(faces, img)
img1 = rectangulos_cara(faces1, img1)
img2 = rectangulos_cara(faces2, img2)
img3 = rectangulos_cara(faces3, img3)
img4 = rectangulos_cara(faces4, img4)

plt.figure(figsize=(13,8))
plt.subplot(2,3,1)
plt.imshow(img)
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1)
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2)
plt.title("Imagen 3")

plt.subplot(2,3,4)
plt.imshow(img3)
plt.title("Imagen 4")

plt.subplot(2,3,5)
plt.imshow(img4)
plt.title("Imagen 5")

plt.show()
```

## Ojos

```python
#cargando las imagenes
img = cv.imread('figuras/people_0.jpg')
img1 = cv.imread('figuras/people_1.jpg')
img2 = cv.imread('figuras/people_2.jpg')
img3 = cv.imread('figuras/people_3.jpg')
img4 = cv.imread('figuras/people_4.jpg')

img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img1_grey = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2_grey = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3_grey = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)
img4_grey = cv.cvtColor(img4, cv.COLOR_BGR2GRAY)
```

```python
#funcion busca jojos
def rectangulo_ojos(eyes, img):
    for (x,y,w,h) in eyes:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,255,255),2)
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
eye_cascade = cv.CascadeClassifier('datos/haarcascade_eye.xml')
```

```python
eyes = eye_cascade.detectMultiScale(img_grey, 1.1 , 5)
eyes1 = eye_cascade.detectMultiScale(img1_grey, 1.1 , 5)
eyes2 = eye_cascade.detectMultiScale(img2_grey, 1.1 , 5)
eyes3 = eye_cascade.detectMultiScale(img3_grey, 1.1 , 5)
eyes4 = eye_cascade.detectMultiScale(img4_grey, 1.1 , 5)
```

```python
#Buscando los ojos en una imagen
img_grey = rectangulo_ojos(eyes, img)
img1_grey = rectangulo_ojos(eyes1, img1)
img2_grey = rectangulo_ojos(eyes2, img2)
img3_grey = rectangulo_ojos(eyes3, img3)
img4_grey = rectangulo_ojos(eyes4, img4)


plt.figure(figsize=(13,8))
plt.subplot(2,3,1)
plt.imshow(img_grey, cmap='gray')
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2_grey, cmap='gray')
plt.title("Imagen 3")

plt.subplot(2,3,4)
plt.imshow(img3_grey, cmap='gray')
plt.title("Imagen 4")

plt.subplot(2,3,5)
plt.imshow(img4_grey, cmap='gray')
plt.title("Imagen 5")

plt.show()
```

## Sonrisa

```python
#cargando las imagenes
img = cv.imread('figuras/people_0.jpg')
img1 = cv.imread('figuras/people_1.jpg')
img2 = cv.imread('figuras/people_2.jpg')
img3 = cv.imread('figuras/people_3.jpg')
img4 = cv.imread('figuras/people_4.jpg')

img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img1_grey = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2_grey = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3_grey = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)
img4_grey = cv.cvtColor(img4, cv.COLOR_BGR2GRAY)
```

```python
#Funcion busca sonrisas
def rectangulo_sonrisa(smiles, img):
    for (x,y,w,h) in smiles:
        cv.rectangle(img,(x,y),(x+w,y+h),(0,255,0),2)
    plt.figure(figsize=(13,8))
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
smile_cascade = cv.CascadeClassifier('datos/haarcascade_smile.xml')
```

```python
smiles = smile_cascade.detectMultiScale(img_grey, 1.3, 20)
smiles1 = smile_cascade.detectMultiScale(img1_grey, 1.3, 20)
smiles2 = smile_cascade.detectMultiScale(img2_grey, 1.3, 20)
smiles3 = smile_cascade.detectMultiScale(img3_grey, 1.3, 20)
smiles4 = smile_cascade.detectMultiScale(img4_grey, 1.3, 20)
```

```python
#Buscando los sonrisas en una imagen
img_grey = rectangulo_sonrisa(smiles, img)
img1_grey = rectangulo_sonrisa(smiles1, img1)
img2_grey = rectangulo_sonrisa(smiles2, img2)
img3_grey = rectangulo_sonrisa(smiles3, img3)
img4_grey = rectangulo_sonrisa(smiles4, img4)


plt.figure(figsize=(13,8))

plt.subplot(2,3,1)
plt.imshow(img_grey, cmap='gray')
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,4)
plt.imshow(img3_grey, cmap='gray')
plt.title("Imagen 3")

plt.subplot(2,3,5)
plt.imshow(img4_grey, cmap='gray')
plt.title("Imagen 4")

plt.show()
```

## Caras con ojos y sonrisas

```python
#cargando las imagenes
img = cv.imread('figuras/people_0.jpg')
img1 = cv.imread('figuras/people_1.jpg')
img2 = cv.imread('figuras/people_2.jpg')
img3 = cv.imread('figuras/people_3.jpg')
img4 = cv.imread('figuras/people_4.jpg')

img_grey = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
img1_grey = cv.cvtColor(img1, cv.COLOR_BGR2GRAY)
img2_grey = cv.cvtColor(img2, cv.COLOR_BGR2GRAY)
img3_grey = cv.cvtColor(img3, cv.COLOR_BGR2GRAY)
img4_grey = cv.cvtColor(img4, cv.COLOR_BGR2GRAY)
```

```python
def rectangulos(faces, eyes, smiles, img):
    for (x,y,w,h) in faces:
        cv.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
        for (x_s,y_s,w_s,h_s) in eyes:
            if( (x <= x_s) and (y <= y_s) and ( x+w >= x_s+w_s) and ( y+h >= y_s+h_s)):
                cv.rectangle(img, (x_s,y_s),(x_s+w_s,y_s+h_s),(255,255,255),3)
        for (x_s,y_s,w_s,h_s) in smiles:
            if( (x <= x_s) and (y <= y_s) and ( x+w >= x_s+w_s) and ( y+h >= y_s+h_s)):
                cv.rectangle(img, (x_s,y_s),(x_s+w_s,y_s+h_s),(0,255,0),3)
    return cv.cvtColor(img, cv.COLOR_BGR2RGB)
```

```python
#Buscando los sonrisas en una imagen
img_grey = rectangulos(faces, eyes, smiles, img)
img1_grey = rectangulos(faces1, eyes1, smiles1, img1)
img2_grey = rectangulos(faces2, eyes2, smiles2, img2)
img3_grey = rectangulos(faces3, eyes3, smiles3, img3)
img4_grey = rectangulos(faces4, eyes4, smiles4, img4)


plt.figure(figsize=(13,8))

plt.subplot(2,3,1)
plt.imshow(img_grey, cmap='gray')
plt.title("Imagen 1")

plt.subplot(2,3,2)
plt.imshow(img1_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,3)
plt.imshow(img2_grey, cmap='gray')
plt.title("Imagen 2")

plt.subplot(2,3,4)
plt.imshow(img3_grey, cmap='gray')
plt.title("Imagen 3")

plt.subplot(2,3,5)
plt.imshow(img4_grey, cmap='gray')
plt.title("Imagen 4")

plt.show()
```
