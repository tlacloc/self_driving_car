```
# Librerias de ayuda
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import cv2 as cv
```

```
def pixel(img):
    img_Ga = cv.GaussianBlur(img,(7,7),0) 
    img_g = cv.cvtColor(img_Ga, cv.COLOR_BGR2GRAY)
    img_r = cv.resize(img_g,(28,28),interpolation=cv.INTER_NEAREST)
    img_i = cv.bitwise_not(img_r )
    return  img_i
```

```
img = cv.imread('/content/drive/My Drive/FashionMnist/panam.jpg')
img_i = pixel(img)
plt.imshow(img_i , cmap="binary")
plt.show()
im1=img_i
```

```
img = cv.imread('/content/drive/My Drive/FashionMnist/sandalias.jpg')
img_i = pixel(img)
plt.imshow(img_i , cmap="binary")
plt.show()
im2=img_i
```

```
img = cv.imread('/content/drive/My Drive/FashionMnist/mochila.jpg')
img_i = pixel(img)
plt.imshow(img_i , cmap="binary")
plt.show()
im3=img_i
```

```
img = cv.imread('/content/drive/My Drive/FashionMnist/playera.png')
img_i = pixel(img)
plt.imshow(img_i , cmap="binary")
plt.show()
im4=img_i
```

```
img = cv.imread('/content/drive/My Drive/FashionMnist/wolverine.jpg')
img_i = pixel(img)
plt.imshow(img_i , cmap="binary")
plt.show()
im5=img_i
```

```
prendas = np.stack((im1, im2,im3, im4, im5))
```
